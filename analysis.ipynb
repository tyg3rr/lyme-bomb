```python
import os, sys
import matplotlib.pyplot as plt
import seaborn as sns
import copy 
from scipy import stats
%matplotlib inline
```

```python
def plot_individuals(df, feature,  sublist=None):
    plt.figure(figsize=(12, 6))
    
    if sublist is None:
        sublist = df.index.unique('ID')
        
    colors =  sns.color_palette(palette='icefire', n_colors=len(sublist), desat=None, as_cmap=False)
    
    for c, idx in zip(colors, sublist):
        indiv_df = df.loc[idx]
        
        ages = indiv_df.index.get_level_values(0).astype('str')
        real_values = indiv_df[feature].values
        plt.plot(ages, real_values, c=c)
        plt.xlabel("Year")
        plt.ylabel(feature)
        plt.title("Evolution for each County of " + feature)
```

```python
plot_individuals(df.set_index(['ID','TIME']), 'Incidence')
```

```python
individual_parameters = pd.DataFrame({'intercept':[], 'slope':[]})

idx = 'Kent'

def compute_individual_parameters(df, idx):
    df_county = df.loc[idx]

    x = df_county.index.get_level_values('TIME').values
    y = df_county['Incidence'].values

    slope, intercept, _, _, _ = stats.linregress(x,y)

    return intercept, slope

individual_parameters.loc[idx] = compute_individual_parameters(df, idx)

individual_parameters
```

```python
for a in df.index.unique('ID'):
    slope, intercept = compute_individual_parameters(df, a)
    individual_parameters.loc[a] = (intercept, slope)

individual_parameters.head()
```
