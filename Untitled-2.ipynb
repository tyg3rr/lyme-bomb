```python
from linearmodels import PanelOLS
```

```python
for var in [
    'Extreme Heat',
    'Extreme Precip.',
    'Total Rainfall',
    'Average Temp (F)'
]:
    dff[(var+'_bin')] = pd.cut(dff[var], bins = 10, right=True)
```

```python
import math
```

```python
exog = [
    'Extreme Heat_bin',
    #'Extreme Precip._bin',
    'Total Rainfall_bin',
    #'Average Temp (F)_bin'
]

mod = PanelOLS(dff['Incidence'], dff[exog], entity_effects=True, time_effects=False)
print(mod.fit())
```

```python
from linearmodels import RandomEffects
```

```python
exog = [
    'Extreme Heat_bin',
    #'Extreme Precip._bin',
    'Total Rainfall_bin',
   # 'Average Temp (F)_bin'
]

mod = RandomEffects(dff['Incidence'], dff[exog])
print(mod.fit())
```
